#Checking for correlation among different factors using heat maps or pair plots.

```python
# Calculate the correlation matrix and creating a heat map
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Greens', linewidths=0.5)
plt.title('Correlation Heat Map')
plt.show()
```

#Insights from heat map
- It is clearly evident that the customers who use treadmill more and run more miles have a positive impact on their health as well as our business. We can encourage people to use the product more which would be mutually beneficial.
- There is also some relationship between education and income indicating that the more educated the person, the more their income. The more their income, the more they are likely to buy medium or advanced equipment.

```python
import numpy as np
# Define criteria and labels for customer profiling
conditions = [
(df['Age'] <= 30) & (df['Income'] <= 40000) & (df['Fitness'] <= 3),
(df['Age'] <= 30) & (df['Income'] <= 40000) & (df['Fitness'] >3),
(df['Age'] <= 30) & (df['Income'] >40000) & (df['Fitness'] <= 3),
(df['Age'] <= 30) & (df['Income'] >40000) & (df['Fitness'] >3),
(df['Age'] >30) & (df['Age'] <= 60) & (df['Income'] <= 40000) & (df['Fitness'] <= 3),
(df['Age'] >30) & (df['Age'] <= 60) & (df['Income'] <= 40000) & (df['Fitness'] >3),
(df['Age'] >30) & (df['Age'] <= 60) & (df['Income'] >40000) & (df['Fitness'] <= 3),
(df['Age'] >30) & (df['Age'] <= 60) & (df['Income'] >40000) & (df['Fitness'] >3),
(df['Age'] >60),
]
labels = [
"Young, Low Income, Low Fitness",
"Young, Low Income, High Fitness",
"Young, High Income, Low Fitness",
"Young, High Income, High Fitness",
"Middle-aged, Low Income, Low Fitness",
"Middle-aged, Low Income, High Fitness",
"Middle-aged, High Income, Low Fitness",
"Middle-aged, High Income, High Fitness",
"Senior Citizens",
]


# Create a new column 'Customer Category' based on criteria
df['Customer Category'] = pd.Categorical(np.select(conditions, labels, default="Other"))
# Visualize the customer categories
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Customer Category', order=labels, palette='viridis')
plt.title('Customer Profile Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

#Recommdendations or Insights

**Gender Based**

- Insight: Female customers tend to purchase the products less when compared to male customers.
- Recommendation: Special discounts and health awareness programs can be given to attract more female customers


**Marital Status Based**

- Insight: Partnered people seem to be conscious about their health in comparison to single people.
- Recommendation: Special discounts can be tailored to attract more single customers to buy the product

**Age Based**

- Insight: People of age between 22 and 37 tend to purchase the product more whereas older people buy the products less and different age groups have different preferences
- Recommendation: Promotions can be tailored specific to different age groups

 **Income Based**

- Insight: Majority of the users lie in the average income category and few in the high income category
- Recommendation: Payment installment plans can be given to customers who buy a premium model compared to their current model

**Education** **Based**

- Insight: It can be inferred from the data that people with good education tend to use the product more
- Recommendation: Health awareness programs can be promoted in order to get more users educated which in turn would lead to more purchase and usage of the product

**Usage Based**

- Insight: Different customers have different usage patterns after buying the product
- Recommendation: Develop strateegies to engage with the customers at regular intervals after their product purchase

**Fitness Scale Based**

- Insight: It is evident from our data that people with high fitness levels tend to buy the advanced model(KP781)
- Recommendation: Emphasize the importance of fitness to every segment of customers in order to upgrade their physical health

**Pricing Strategies**

- Recommendation: Consider tweaking the prices by tailoring different pricing strategies based on income of users

**Customer segmentation**

- Recommendation: Tailor workout strategies for different customer segments based on their usage patterns and fitness levels in order to improve product engagement which can ultimately lead to their product upgrade

**Product Improvement**

- Insight: The data suggests that there is more room for improvement in the product to attract more customers to buy our product
- Recommendation: While giving importance to pricing strategies and marketing campaigns, give equal importance to product development too.
